# NIS Plot
Based on the chart below, the values I choose for longitudinal and yaw acceleration noise are acceptable. These values are 2 m/s^2 and pi/2 rad/s^2. The majority of plotted values fall within the suggested 95% confidence bounds for NIS measurements with three degrees of freedom.

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv = np.genfromtxt ('src/nis.csv', delimiter=",")
csv_edited = np.delete(csv, [0, csv.size - 1])
```

```python
plt.figure()
plt.plot(csv_edited)
plt.hlines([0.352,7.815], 0, csv_edited.size)
plt.title("NIS")
plt.xlabel("Record")
plt.ylabel("Value")
plt.show()
```
